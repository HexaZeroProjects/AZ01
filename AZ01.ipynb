```python
!wget 'https://raw.githubusercontent.com/HexaZeroProjects/AZ01/main/dz.csv' -O dz.csv
column_names = ['Name', 'City', 'Salary']
df2 = pd.read_csv('dz.csv', header=None, names=column_names, delimiter=',')
df2
```

```python
# Преобразуем столбец 'Salary' в числовой формат, игнорируя ошибки, если есть пустые значения
df2['Salary'] = pd.to_numeric(df2['Salary'], errors='coerce')

# Вычисляем среднее значение по зарплате, игнорируя пустые значения
average_salary = df2['Salary'].mean()
print("Средняя зарплата:", average_salary)
```
